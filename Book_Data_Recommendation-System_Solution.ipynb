###### 2. using Eucledian Distance as a metric for UBCF

```python
book_data.columns
```

```python
book_data_1 = book_data.drop(columns=['Unnamed: 0'])
```

```python
book_data_1
```

```python
from sklearn.metrics import pairwise_distances
```

```python
euclidean_pivot_table = pd.pivot_table(data=book_data_1,values='Book.Rating',index='User.ID',columns='Book.Title').fillna(0)
euclidean_pivot_table
```

```python
book_data_1['User.ID'].unique()
```

```python
euclidean_pivot_table.index =book_data_1['User.ID'].unique()
euclidean_pivot_table
```

```python
euclidean_pivot_table.values
```

```python
euclidean_ubcf = pairwise_distances(X=euclidean_pivot_table.values,metric='euclidean')
euclidean_ubcf
```

```python
book_data_1['User.ID'].unique()
```

```python
ubcf_euclidean_metric = pd.DataFrame(data=euclidean_ubcf)
ubcf_euclidean_metric.index = book_data_1['User.ID'].unique()
ubcf_euclidean_metric.columns=book_data_1['User.ID'].unique()
ubcf_euclidean_metric.round(2)
```

###### 3. Using Cosine as a metric for UBCF

```python
cosine_pivot_table = pd.pivot_table(data=book_data_1,values='Book.Rating',index='User.ID',columns='Book.Title').fillna(0)
cosine_pivot_table
```

```python
cosine_pivot_table.index = book_data_1['User.ID'].unique()
cosine_pivot_table
```

```python
cosine_pivot_table.values
```

```python
cosine_ubcf = 1-pairwise_distances(X=cosine_pivot_table.values,metric='cosine')
cosine_ubcf
```

```python
book_data_1['User.ID'].unique()
```

```python
cosine_ubcf_df         = pd.DataFrame(data=cosine_ubcf)
cosine_ubcf_df.index   = book_data_1['User.ID'].unique()
cosine_ubcf_df.columns = book_data_1['User.ID'].unique()
cosine_ubcf_df.round(2)
```

###### Filtering top 50 data

```python
first_50_user = cosine_ubcf_df.iloc[:50,:50]
first_50_user
```
